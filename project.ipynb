#kNN → weak recall for churn (0.51)

#Random Forest → recall dropped further (0.49)

#XGBoost (current) → ❌ completely failed (predicts only “No”)
# Logistic Regression is selected as the final model because:
# 1. The dataset is imbalanced, and recall for churned customers is critical
# 2. Logistic Regression achieved the highest recall and F1-score for the 'Yes' (churn) class
# 3. The model is interpretable and suitable for business decision-making
# 4. XGBoost failed due to class imbalance, and Random Forest underperformed on churn recall


```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report,confusion_matrix,roc_auc_score
```

```python
final_model = LogisticRegression(
    class_weight='balanced',
    max_iter=1000,
    random_state = 42
)
final_model.fit(x_train_sc,y_train)
```

```python
y_prob = final_model.predict_proba(x_test_sc)[:,1]
print(y_prob)
```

```python
# threshold tuning (lower than 0.5 to catch churn)
y_pred_final = (y_prob >= 0.35).astype(int)
```

```python
print(accuracy_score(y_test,y_pred_final))
```

```python
print("Final Logistic Regression Model")
print(classification_report(y_test, y_pred_final))

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_final))

print("ROC-AUC Score:", roc_auc_score(y_test, y_prob))
```

clear  conclusion -
The objective of this project was to identify customers at high risk of churn so that proactive retention strategies could be applied. Using customer demographic, service usage, and account information, a Logistic Regression model was developed and optimized to prioritize the detection of churned customers.

Due to the imbalanced nature of the dataset, accuracy alone was not sufficient for evaluation. The model was therefore optimized for recall of churned customers, as failing to identify a customer who is about to leave has a higher business cost than incorrectly flagging a loyal customer.

The final model achieved:

92% recall for churned customers

ROC-AUC score of 0.85, indicating strong class separation ability

This means the model is able to correctly identify the majority of customers who are likely to churn, enabling the business to intervene early with retention offers, improved service, or personalized engagement.

Although the model produces some false positives, this trade-off is acceptable in a churn prediction scenario, as the cost of offering retention incentives to some loyal customers is significantly lower than losing high-value customers.

Business Impact-

Supports data-driven retention strategies

Helps reduce customer attrition and revenue loss

Provides an interpretable model suitable for business decision-making

Recommendations for the Business-

Target customers predicted as high churn risk with retention campaigns

Focus on customers with short tenure and flexible contracts, as they show higher churn tendency

Continuously retrain the model as customer behavior and service offerings evolve
